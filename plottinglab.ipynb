```python
#Standard deviation, Standard Error, Confidence interval
import numpy as np
from scipy.stats import t
from matplotlib import pyplot as plt



Y = np.random.randint(1, 11, 11) #Generates 10 random datapoints


n = Y.size #Sample Size


Y_mean = np.mean(Y) #Calculates Mean


Y_std = np.std(Y) # Calculates standard deviation


Y_se = Y_std / np.sqrt(n) # Calculates Standard error

# 75% Confidence Interval

dof = n - 1         # degrees of freedom
alpha = 1.0 - 0.75
conf_interval = t.ppf(1-alpha/2., dof) * Y_std*np.sqrt(1.+1./n)

fig = plt.gca()
plt.errorbar(1, Y_mean, yerr=Y_std, fmt='-o')
plt.errorbar(2, Y_mean, yerr=Y_se, fmt='-o')
plt.errorbar(3, Y_mean, yerr=conf_interval, fmt='-o')

plt.xlim([0,4])
plt.ylim(Y_mean-conf_interval-2, Y_mean+conf_interval+2)

# axis formatting
fig.axes.get_xaxis().set_visible(False)
fig.spines["top"].set_visible(False)
fig.spines["right"].set_visible(False)
plt.tick_params(axis="both", which="both", bottom="off", top="off",
                labelbottom="on", left="on", right="off", labelleft="on")

plt.legend(['Standard Deviation', 'Standard Error', 'Confidence Interval'],
           loc='upper left',
           numpoints=1,
           fancybox=True)

plt.ylabel('random variable')
plt.title('10 random values in the range 1-11')

plt.show()
```

```python
#Error bars
import matplotlib.pyplot as plt

# input data
mean_values = [3, 5, 7]
variance = [0.1, 0.4, 0.5]
bar_labels = ['Value 1', 'Value 2', 'Value 3']

fig = plt.gca()


x_pos = list(range(len(bar_labels)))
plt.bar(x_pos, mean_values, yerr=variance, align='center', alpha=0.5)  #Plotting bars


max_y = max(zip(mean_values, variance)) # returns a tuple, here: (3, 5)
plt.ylim([0, (max_y[0] + max_y[1]) * 1.1])

# set axes labels and title
plt.ylabel('variable y')
plt.xticks(x_pos, bar_labels)
plt.title('Bar plot with error bars')

# axis formatting
fig.axes.get_xaxis().set_visible(False)
fig.spines["top"].set_visible(False)
fig.spines["right"].set_visible(False)
plt.tick_params(axis="both", which="both", bottom="off", top="off",
                labelbottom="on", left="on", right="off", labelleft="on")

plt.show()
```

```python
#Pie-chart
from matplotlib import pyplot as plt
import numpy as np

plt.pie(
    (5,10,15,20),
    labels=('Workers','Materials', 'Teachers', 'Students'),
    shadow=True,
    colors=('red', 'yellow', 'pink', 'green'),
    explode=(0,0.10,0.10,0.10),
    startangle=90,    # rotate counter-clockwise by 90 degrees
    autopct='%1.1f%%',# displays fraction as percentage
    )
plt.legend(fancybox=True)
plt.axis('equal')     # plot pyplot as circle
plt.tight_layout()
plt.show()
```
